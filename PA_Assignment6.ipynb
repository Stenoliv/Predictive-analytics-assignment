```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from pmdarima import auto_arima

# Load Excel data
file_path = "/content/Data for Assignment SARIMA and LinRegression (1).xlsx"
exports_df = pd.read_excel(file_path, sheet_name='Exports')
population_df = pd.read_excel(file_path, sheet_name='Population')

# Create datetime index
def create_datetime_index(df):
    return pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'], format='%Y-%b')

exports_df['Date'] = create_datetime_index(exports_df)
population_df['Date'] = create_datetime_index(population_df)
exports_df.set_index('Date', inplace=True)
population_df.set_index('Date', inplace=True)

# Extract time series
exports_series = exports_df['Exports by Atlantis in Million USD']
population_series = population_df['Population at end of this month']

# Decomposition of data
exports_decomp = seasonal_decompose(exports_series, model='additive', period=12)
population_decomp = seasonal_decompose(population_series, model='additive', period=12)

exports_decomp.plot()

plt.show()

population_decomp.plot()

plt.show()

# ACF/PACF Plots
plot_acf(exports_series, lags=24)
plt.title("ACF - Exports")
plt.show()

plot_pacf(exports_series, lags=24)
plt.title("PACF - Exports")
plt.show()

# SARIMA Forecasting
sarima_model_exports = auto_arima(exports_series, seasonal=True, m=12, trace=True, suppress_warnings=True)
exports_forecast = sarima_model_exports.predict(n_periods=6)

sarima_model_population = auto_arima(population_series, seasonal=True, m=12, trace=True, suppress_warnings=True)
population_forecast = sarima_model_population.predict(n_periods=6)

# Linear Regression Forecasting
def prepare_lr_features(series):
    df = pd.DataFrame({'y': series})
    df['month'] = df.index.month
    df['time'] = np.arange(len(df))
    ohe = OneHotEncoder(drop='first', sparse_output=False)

    month_encoded = ohe.fit_transform(df[['month']])
    month_df = pd.DataFrame(month_encoded, index=df.index, columns=[f'Month_{i}' for i in range(1, 12)])
    return pd.concat([df[['y', 'time']], month_df], axis=1), ohe

# Prepare features for Exports
exports_lr_df, ohe = prepare_lr_features(exports_series)
X_exports = exports_lr_df.drop(columns='y')
y_exports = exports_lr_df['y']
lr_exports = LinearRegression().fit(X_exports, y_exports)

# Prepare future months
future_time = np.arange(len(exports_series), len(exports_series)+6)
future_months = [(exports_series.index[-1] + pd.DateOffset(months=i)).month for i in range(1, 7)]
future_months_df = pd.DataFrame({'month': future_months})
future_months_encoded = ohe.transform(future_months_df)
future_months_encoded = pd.DataFrame(future_months_encoded, columns=X_exports.columns[1:])
X_future_exports = pd.DataFrame({'time': future_time})
X_future_exports = pd.concat([X_future_exports, future_months_encoded], axis=1)
lr_exports_forecast = lr_exports.predict(X_future_exports)

# Repeat for Population
population_lr_df, ohe2 = prepare_lr_features(population_series)
X_pop = population_lr_df.drop(columns='y')
y_pop = population_lr_df['y']
lr_population = LinearRegression().fit(X_pop, y_pop)

future_months = [(population_series.index[-1] + pd.DateOffset(months=i)).month for i in range(1, 7)]
future_months_df = pd.DataFrame({'month': future_months})
future_months_encoded = ohe2.transform(future_months_df)
future_months_encoded = pd.DataFrame(future_months_encoded, columns=X_pop.columns[1:])
X_future_pop = pd.DataFrame({'time': future_time})
X_future_pop = pd.concat([X_future_pop, future_months_encoded], axis=1)
lr_population_forecast = lr_population.predict(X_future_pop)

# Plot and Compare
future_dates = pd.date_range(start=exports_series.index[-1] + pd.DateOffset(months=1), periods=6, freq='MS')

# Exports
plt.figure(figsize=(12, 5))
plt.plot(exports_series, label='Historical Exports')
plt.plot(future_dates, exports_forecast, label='SARIMA Forecast', marker='o')
plt.plot(future_dates, lr_exports_forecast, label='Linear Regression Forecast', marker='x')
plt.legend()
plt.title('Exports Forecast Comparison')
plt.show()

# Population
plt.figure(figsize=(12, 5))
plt.plot(population_series, label='Historical Population')
plt.plot(future_dates, population_forecast, label='SARIMA Forecast', marker='o')
plt.plot(future_dates, lr_population_forecast, label='Linear Regression Forecast', marker='x')
plt.legend()
plt.title('Population Forecast Comparison')
plt.show()
```
